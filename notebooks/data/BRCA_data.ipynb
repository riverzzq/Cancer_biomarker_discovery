```python
import pandas as pd
import numpy as np
```

```python
clinical = pd.read_csv("BRCA-Clinical.tsv", sep="\t", usecols = ["Sample ID", "Subtype"], index_col = 0)
clinical = clinical.dropna()
```

```python
output_path = "BRCA-Clinical.tsv"
clinical.to_csv(output_path, sep="\t")
```

```python
expr = pd.read_csv("BRCA-RNASeq.tsv", sep = "\t", index_col=0)
expr = expr.T
expr.columns.name = "Sample ID"
expr
```

```python
expr.index = expr.index.str[:15]
common_ids = expr.index.intersection(clinical.index)
```

```python
expr_aligned = expr.loc[common_ids]
clinical_aligned = clinical.loc[common_ids]
```

```python
expr_with_subtype = expr_aligned.join(clinical_aligned["Subtype"])
```

```python
expr_with_subtype
```

```python
output_path = "BRCA-data.tsv"
expr_with_subtype.to_csv(output_path, sep="\t")
```
